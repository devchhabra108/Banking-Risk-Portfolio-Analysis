```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("/content/clients_clean.csv")
df.head(5)
```

```python
df.shape
```

```python
df.info()
```

```python
# Generate descriptive statistics for the dataframe
df.describe()
```

```python
bins = [0, 100000, 300000, float('inf')]
labels = ['Low', 'Medium', 'High']
df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels ,right = False)
```

```python
df['Income Band'].value_counts().plot(kind = 'bar')
```

```python
# Examine the distribution of unique categories in categorical columns
categorical_cols = df[["BRId", "GenderId","IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"]].columns
for col in categorical_cols:
    print(f"Value counts{col}:")
    display(df[col].value_counts())
```

```python
for i, predictor in enumerate(df[["BRId", "GenderId","IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"]]):
     plt.figure(i)
     sns.countplot(data = df, x = predictor, hue = "Nationality")
```

```python
# Histplot of value counts for different occupation
for col in categorical_cols:
  if col == "Occupation":
    continue
  plt.figure(figsize=(8,4))
  sns.histplot(df[col])
  plt.title('Histogram of occupation count')
  plt.xlabel(col)
  plt.ylabel("Count")
  plt.show()
```

```python
# Numerical Analysis
```

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

#Univariate Analysis and Visualization
plt.figure(figsize=(15,10))
for i, col in enumerate(numerical_cols):
  plt.subplot(4,3,i+1)
  sns.histplot(df[col], kde = True)
  plt.title(col)
plt.show()
```

Heatmaps

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

correlation_matrix = df[numerical_cols].corr()
plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt =".2f")
plt.title('Correlation Heatmap')
plt.show
```

**Insights Of EDA**



1.   The strongest positive correlation occur among "Bank Deposits" with "Checking Accounts" and "Foreign Currency Account" indicating that customer who maintain high balances in one account type often hold substantial amount/ funds across other accounts as well.

2. The dataset consists of 3,000 customer records and 25 attributes with no missing values, ensuring that the data was complete, consistent, and suitable for further analysis and visualization.

3. Customer demographics and banking characteristics were analyzed across Gender, Nationality, Occupation, Banking Relationship, and Income Bands. This segmentation provided a clear understanding of the customer portfolio and supported comparative analysis across different customer groups.

4. The distributions of major financial variables—including Estimated Income, Bank Loans, Bank Deposits, Business Lending, Savings Accounts, Checking Accounts, Credit Card Balance, Foreign Currency Accounts, and Superannuation Savings—revealed significant variation among customers, highlighting the diversity of financial profiles within the banking portfolio.

5. Most of the remaining financial variables exhibited weak to moderate correlations, suggesting that each banking product contributes unique information about customer financial behavior rather than providing redundant insights.
